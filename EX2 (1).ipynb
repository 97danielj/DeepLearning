```python
from sklearn.datasets import *
t=load_boston()#데이터 로드
x=t.data
x
```

```python
y=t.target#정답
y
```

```python
from sklearn.linear_model import LinearRegression
m=LinearRegression()#모델 생성
m.fit(x,y)#학습
#Y=WX->Y=10X
out_d=m.predict(t.data)
plt.scatter(y,out_d)
plt.show()
```
